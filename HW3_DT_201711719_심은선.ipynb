## 1. Load Dataset, train/test split

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

datasets = load_iris()
# datasets
```

```python
X = datasets['data']
y = datasets['target']
features = datasets['feature_names']
classes = datasets['target_names']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2020)
```

```python
print('The number of train set:', len(X_train))
print('The number of test set:', len(X_test))
```

## 2. Fitting DT & visualization

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(X_train, y_train)
```

```python
import graphviz

dot_data = tree.export_graphviz(clf, out_file=None, filled=True, feature_names =features, class_names=classes)
graph = graphviz.Source(dot_data)
```

```python
graph
```

## 3. 성능 평가

```python
pred_y = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print('Accuracy of DT:', accuracy_score(y_test, pred_y))
```
